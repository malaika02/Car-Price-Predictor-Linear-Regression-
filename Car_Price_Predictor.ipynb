```python
#Clean data
car = df.copy()
car = car[car['year'].str.isnumeric()]  #sirf wo e rows de ga jin mn car ki value numeric ho gi
car
```

```python
#remove nan values from fuel type
car= car[~car['fuel_type'].isna()]
car.info()
```

### **Model**

```python
#extract features
x = car.drop(columns = ['Price'])
y = car['Price']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y, test_size=0.2)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import make_column_transformer
from sklearn.pipeline import make_pipeline
```

```python
ohe = OneHotEncoder()
ohe.fit(x[['name','company','fuel_type']])
```

```python
column_trans = make_column_transformer((OneHotEncoder(categories=ohe.categories_),['name','company','fuel_type']),
                                       remainder = 'passthrough')
```

```python
lr = LinearRegression()
```

```python
pipe = make_pipeline(column_trans,lr) #jo input data dety hn hm pipeline ko wo phly column_trans pa jye ga onehotencoder
#lgy ga saari columns transform hon gi us k bad sari transform column along with previous column will be given to lr
```

```python
pipe.fit(x_train,y_train)
```

```python
y_pred = pipe.predict(x_test)
y_pred
```

```python
r2_score(y_test,y_pred)
```

```python
pipe.predict(pd.DataFrame([['Maruti Suzuki Swift','Maruti',2019,100,'Petrol']],columns = ['name','company','year','kms_driven','fuel_type']))
```
